```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re

# 엑셀 파일 이름
file_name = '2020 ~ 2021 수업시간표.xlsx'

# iconv 명령 실행
os.system(f"iconv -f ISO-8859-1 -t UTF-8 {file_name} > 2020 ~ 2021 수업시간표.xlsx")
    
# 데이터 불러오기
df = pd.read_excel('2020 ~ 2021 수업시간표.xlsx')
df
```

```python
# 강의실 별로 몇 개의 강의가 있는지 출력
print(df['강의실'].value_counts())

# 강의실 별 강의 수 시각화
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(100,5))
sns.countplot(x='강의실', data=df)
plt.title('강의실 별 강의 수')
plt.show()
```

```python
# 교수님 별 강의 수 출력
print(df['교수명'].value_counts())

# 교수님 별 강의 수 시각화
plt.figure(figsize=(80,5))
sns.countplot(x='교수명', data=df)
plt.xticks(rotation=90)
plt.title('교수님 별 강의 수')
plt.show()
```

```python
# 강의 요일별 강의 수 출력
print(df['강의시간'].value_counts())

# 강의 요일별 강의 수 시각화
plt.figure(figsize=(70,5))
sns.countplot(x='강의시간', data=df)
plt.title('강의시간 요일별 강의 수')
plt.show()
```
