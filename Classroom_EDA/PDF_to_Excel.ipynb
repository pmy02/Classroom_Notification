```python
import pandas as pd
import numpy as np

# 엑셀 파일에서 모든 시트를 불러옴
df = pd.read_excel('2020 수업시간표 - pdf 변환.xlsx', sheet_name=None)
df
```

```python
combined_df = pd.concat(df.values(), ignore_index=True)
combined_df = combined_df[['교과목명', '교수명', '강의시간/강의실']]
combined_df
```

```python
# 결측값이 있는 행 삭제
combined_df.dropna(inplace=True)

# 강사/교수미정 학과 제거
combined_df = combined_df[~combined_df['교수명'].str.contains('강사미정')]
combined_df = combined_df[~combined_df['교수명'].str.contains('교수미정')]

# 교수명에서 학과 제거
combined_df['교수명'] = combined_df['교수명'].str.replace(r'\(.*?\)', '', regex=True)
    
# 강의시간과 강의실 데이터 분리
combined_df['강의실'] = combined_df['강의시간/강의실'].str.extract('\[(.*)\]')[0]
combined_df['강의시간/강의실'] = combined_df['강의시간/강의실'].str.replace(r'\[.*?\]', '', regex=True)
combined_df = combined_df.rename(columns={'강의시간/강의실': '강의시간'})

# 값이 없는 행 결측값으로 변경
combined_df['강의시간'].replace(' ', np.nan, inplace=True)
combined_df['강의실'].replace(' ', np.nan, inplace=True)

# 결측값이 있는 행 삭제
combined_df.dropna(inplace=True)

combined_df
```

```python
# 저장하기
combined_df.to_excel('2020 수업시간표 - 시트 통합.xlsx', index=False)
```
